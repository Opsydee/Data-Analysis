# Visualization

- Pandas: Built on Matplotlib, simplifies quick plotting directly from DataFrames 
- Seaborn: Built on Matplotlib simplifies complex statistics visalization and improves the statistics of the plots.
- Matplotlib: The foundation of both pandas and seaborn visualization. it provides the core plotting functions., but with more flexibility at the cost of increased complexity

When to use each library

- Pandas: when you need quick, basic visualizations dirctly from your Dataframe.
    
- Matplotlib: when you need fine control and customization of plots or when creating more complex visualizations.

- Seaborn: when you want aesthetically pleasing plots with less code andd when you are focusing on statistical relationships in the data  

## Most Commonly used Visuals 
-Bar Chart: Categorical comparison using rectangular bars
   - when to use : Comparing different groups or categories (sales by region, products by popularity)
   - Use for: Discrete Categories when you havve a few categories 
- Histogram: Distribution of a numeric variables using bars 
   - when to use: To observe the frequency distribution of data(e.g. age distribution,income ranges )
   - Use for:Understand the spread or skewness of data
- Line chart: Usually for time series data 
   - when to use: To show trends or changes over time (e.g. stock prices over Months)
   - Use For: Tracking changes over a continuous intervals(Like time)
- Scatter Plot: Points Plotted on two axes to display relationships between two variables.
   - When to use: To show correlation or relationships between two numeric variables (e.g Sales vs Profit)
   - Use For : identifying relationships clutters and outlier 
- Pie Chart: circular chart divided into slices to represent proportions 
   - when to use : To show parts of a whole (e.g market share,budget allocating)
   - Use For: Simple proportional comparisons avoid with many categories.
- Boxplot: Display distribution quartikes and outliers of data
   - When to use: To compare distribution across categories (e.g profits across different product lines)
   - Use For: Visualizing spread, outliers and comparing distributions particular with statistical data
